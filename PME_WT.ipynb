```python
import warnings
warnings.filterwarnings('ignore')
from LogDensity import LogDensity
from Mixed import Mixed
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

# Log density method

```python
sim = LogDensity(xrange=(-np.pi,np.pi),m=3,nx=800,dt=1e-3,dim=1,plot=0,IC='waitingTime',plotEvery=1)
plt.figure()
plt.plot(sim.pnts_x,sim.getPointsRhoex(),'-k',label = r'$t=0$')
plt.xlabel(r'$x$'); plt.ylabel(r'$\rho$')
sim.simulate(0.125)
plt.plot(sim.pnts_x,sim.rhok.vec,'--r',label = r'$t=0.125$')
sim.simulate(0.25)
plt.plot(sim.pnts_x,sim.rhok.vec,'--b',label = r'$t=0.15$')
plt.legend(frameon=0,loc='upper left',fontsize=10)
plt.show()
```

## Density plots

m=2

```python
nxList = [200*2**j for j in range(7)]
dt = 1e-3
m=2
time = [(i+1)*dt for i in range(0,int(.2/dt))]
DEN = []
for nx in nxList:
    b = LogDensity(xrange=(-np.pi,np.pi),m=m,nx=nx,dt=dt,dim=1,plot=0,IC='waitingTime')
    denxR = []
    for t in time:
        b.simulate(t)
        denxR.append(b.rhok.vec[b.initialxRNode])
    DEN.append(denxR)
```

```python
fig, ax = plt.subplots()
ax.tick_params(direction="in",which='both',axis='both',
                          bottom=True, top=True, left=True, right=True)
ax.set_ylabel('Density at node $x_R^0$',fontsize=12)
ax.set_xlabel('$t$',fontsize=12)
ax.set_xlim([0, .2]); ax.set_ylim([0, .004])
ax.minorticks_on()
theoreticalWT = 1/2/(m+1)
ax.axvline(x=theoreticalWT,linestyle='-',color='black',label='Theoretical Waiting Time')
for i in range(len(DEN)):
    ax.plot(time,DEN[i],'--',label='N='+str(nxList[i]))
ax.legend(frameon=0,loc='upper left')
plt.show()
```
