```python
from LogDensity import LogDensity
from Mixed import Mixed
import numpy as np
from IPython.display import display, HTML
import warnings
warnings.filterwarnings('ignore')
display(HTML("<style>.container { width:100% !important; }</style>"))
import matplotlib.pyplot as plt
%matplotlib notebook
```

# 1D plots

Log density

```python
sim = LogDensity(xrange=(-10,10),s0=3,m=3,nx=200,dt=5e-2,dim=1,plot=1,IC='BB',printTime=True,plotInterface=0)
sim.plot = False
sim.simulate(1)
plt.plot(sim.pnts_x,sim.rhok.vec,'bs',markersize=4,label=r'$\rho(1,x)$')
plt.plot(sim.pnts_x,sim.getPointsRhoex(),'-',color='gray',label=r'$\rho_{exact}(1,x)$')
handles, labels = plt.gca().get_legend_handles_labels()
labels[0],labels[1] = r'$\rho(0,x)$',r'$\rho_{exact}(0,x)$'
labels[1],labels[2] = labels[2],labels[1]
handles[1],handles[2] = handles[2],handles[1]
plt.legend(handles, labels,fontsize=10,frameon=0,loc='upper left')
plt.show()
```

Mixed

```python
sim = Mixed(xrange=(-10,10),s0=3,m=3,nx=200,dt=5e-2,dim=1,plot=1,IC='BB',printTime=True,plotInterface=0)
sim.plot = False
sim.simulate(1)
plt.plot(sim.pnts_xC,sim.rhoh.vec,'s',color='blue',markersize=4,label=r'$\rho(1,x)$')
plt.plot(sim.pnts_x,sim.getPointsRhoex(),'-',color='gray',label=r'$\rho_{exact}(1,x)$')
handles, labels = plt.gca().get_legend_handles_labels()
labels[0],labels[1] = r'$\rho(0,x)$',r'$\rho_{exact}(0,x)$'
labels[1],labels[2] = labels[2],labels[1]
handles[1],handles[2] = handles[2],handles[1]
plt.legend(handles, labels,fontsize=10,frameon=0,loc='upper left')
plt.show()
```
